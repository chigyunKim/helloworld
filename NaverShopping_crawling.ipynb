```python
import selenium
from selenium import webdriver 
from selenium.webdriver.common.keys import Keys #엔터 입력이나 스크롤 제어 등등
import time
import csv #csv 파일 생성, 입력 등등
```

```python
driver = webdriver.Chrome() #크롬드라이버 실행

driver.get('https://www.naver.com') #네이버 홈페이지 열기
driver.implicitly_wait(10) #로딩 완료까지 대기

driver.find_element_by_css_selector('a.nav.shop').click() #쇼핑 메뉴 클릭
time.sleep(2) #로딩 완료까지 대기
```

```python
#검색칸 선택
search = driver.find_element_by_css_selector('input._searchInput_search_text_3CUDs')
search.click()

#검색어 입력
search.send_keys('아이폰 13')
search.send_keys(Keys.ENTER)
```

```python
#상품페이지가 스크롤 하면 추가 생성되는 구조이므로, 스크롤을 끝까지 한 뒤 크롤링 하기 위한 코드
#스크롤 전체 높이
before_h = driver.execute_script('return window.scrollY') #자바스크립트 명령어를 사용 가능하게 해주는 코드

#상품 몇개 있는지 F12 화면에서 확인하는 방법.
#F12화면에서 "Elements" 선택 후, 크롤링 대상 class 찾은 다음, class명칭 복사한 뒤, ctrl+F 누르고 입력하면됨


#무한스크롤
while True:
    driver.find_element_by_css_selector('body').send_keys(Keys.END)#맨 아래로 스크롤을 내린다.    
    time.sleep(2) #스크롤 사이 페이지 로딩시간
    after_h = driver.execute_script('return window.scrollY') #스크롤 후 높이
    
    if after_h == before_h: #스크롤 끝까지 내리게 되면, 이전 높이와 같아짐. 이때 break 걸림.
        break
    before_h = after_h

#csv파일생성
f = open(r'C:\Users\chigy\Practice\NaverShoppingData.csv', 'w', encoding='CP949', newline='') #csv파일 생성
#encoding='CP949' 입력 안하면 글자깨진다고 함.실험해볼 것. newline='' 입력 안하면 줄바꿈이 이상하게 된다고 함.
csvWriter = csv.writer(f)


#상품정보 div 정보
items = driver.find_elements_by_css_selector('.basicList_info_area__TWvzp') #전체 상품 div 정보를 담기에 element"s"임.

for item in items:
    name = item.find_element_by_css_selector('.basicList_title__VfX3c').text #상품명만 가져옴
    try:
        price = item.find_element_by_css_selector('.price_num__S2p_v').text #가격만 가져옴
    except:
        price = "판매중단" #가격 정보를 가져올 수 없다면, "판매중단"으로 입력 처리할 것.
    link = item.find_element_by_css_selector('.basicList_title__VfX3c > a').get_attribute('href')
    #.basicList_title__VfX3c의 하위에 있으므로 ">a"로 붙임. 
    print(name, price, link)
    csvWriter.writerow([name,price,link]) #writerow는 한행씩 덧붙이겠다는 의미
    
f.close() #입력 끝난 후 항상 닫아줄 것.
```
